```python
import math
# Exercice 1 (CM) : Classe Point2D
class Point2D:
    def __init__(self, x=0, y=0):
        self._x = x
        self._y = y
        self._r = math.sqrt(self._x**2+self._y**2)
        self._theta = math.atan2(self._y,self._x)

    def _get_x(self):
        return self._x
    
    def _get_y(self):
        return self._y

    def _get_r(self):
        return self._r

    def _get_theta(self):
        return self._theta

    def _set_x(self,new_x ):
        self._x  =  new_x

    def _set_y(self,new_y ):
        self._y  =  new_y

    def _set_r(self,new_r):
        self._r  =  new_r

    def _set_theta(self,new_theta ):
        self._theta  =  new_theta

    def cartesian2Polar(self) :
        self._r = math.sqrt(self._x**2+self._y**2)
        self._theta = math.atan2(self._y,self._x)

    def polar2Cartesian(self):
        self._x = self._r*math.cos(self._theta)
        self.y = self._r*math.sin(self._theta)

    def print(self):
        print("Cartesian Coordinates: %f %f"%(self._x,self._y))
        print("Polar Coordinates: %f %f"%(self._r,self._theta))

    x = property(_get_x, _set_x)
    y = property(_get_y, _set_y)
    r = property(_get_r, _set_r)
    theta = property(_get_theta, _set_theta)

    
p = Point2D(1,1)
p.print()
p.cartesian2Polar()
p.print()
p.r = 2
p.theta = math.pi/4
p.polar2Cartesian()
p.print()
```

```python
# Exercise 2: Robot class

class Robot:
    def __init__(self, name: str = 'John', pos: Point2D = Point2D(0, 0)):
        self._name = name
        self._pos = pos
        
    def _get_name(self):
        return self._name
    def _get_pos(self):
        return self._pos
    
    def _set_name(self,new_name ):
        self._name  =  new_name
    def _set_pos(self,new_pos ):
        self._pos  =  new_pos
        
    name = property(_get_name, _set_name)
    pos = property(_get_pos, _set_pos)

rob = Robot()
print(rob.name)
print(rob._pos._get_x(), rob._pos._get_y())
print(rob._get_pos())
     
```

```python
# Exercise 3: Board Class
class Board:
    def __init__(self,  grid: list[list[bool]],robots: list[Robot]= [], exit: Point2D = Point2D(0,0)) -> None:
        self._grid = grid
        self._robots = robots
        self._exit = exit
        
    def _get_grid(self):
        return self._grid
    def _get_robots(self):
        return self._robots

    def _set_grid(self,new_grid):
        self._grid  =  new_grid
    def _set_robots(self,new_robots):
        self._robots  =  new_robots
        
    def add_robot(self, robot: Robot):
        self._robots.append(robot)    
    def remove_robot(self, robot_idx: int):
        self._robots.pop(robot_idx)
        
    def moveRobot(self, robot_idx: int, end: Point2D):
        # Find if the robot can be move through the grid to the end point
        # Grid is a 2d matrix of boolean, True if the cell is free, False if there is an obstacle
        # Robot can move in 4 directions: up, down, left, right
        
        robot = self._robots[robot_idx]
        start = robot.pos
        if not self._grid[end._y][end._x]:
            print("Cannot move to the end point, there is an obstacle")
            return False
        if start._x == end._x:
            step = 1 if start._y < end._y else -1
            for y in range(start._y, end._y, step):
                if not self._grid[y][start._x]:
                    print("Cannot move to the end point, there is an obstacle")
                    return False
        elif start.y == end._y:
            step = 1 if start._x < end._x else -1
            for x in range(start._x, end._x, step):
                if not self._grid[start._y][x]:
                    print("Cannot move to the end point, there is an obstacle")
                    return False
        else:
            print("Cannot move to the end point, not in the same row or column")
            return False
        robot.pos = end
        return True
        
        
    grid = property(_get_grid, _set_grid)
    robots = property(_get_robots, _set_robots)
    
board = Board([[True, True, True, True, True],
               [True, False, False, False, True],
               [True, True, True, False, True],
               [True, False, True, True, True],
               [True, True, True, False, True]])
robot1 = Robot("R1", Point2D(0,0))
robot2 = Robot("R2", Point2D(4,4))
board.add_robot(robot1)
board.add_robot(robot2)
print(board.robots)
print(board.moveRobot(0, Point2D(0,4)))
print(board.robots[0].pos._x, board.robots[0].pos._y)
```

```python
# Exercise 4: Alice and Bob
# Init Alice and Bob. We can see that Alice is at (10, 2) and Bob is at (6.5, 6)
alice = Robot("Alice", Point2D(10, 2))
bob = Robot("Bob", Point2D(6.5, 6))


# Init a tray of 11 rows and 11 columns with walls and 1 entrance at (0, 1)

tray = [[False]*11 for _ in range(11)]
for i in range(11):
    for j in range(11):
        if i == 0 or i == 10 or j == 0 or j == 10:
            tray[i][j] = False
        else:
            tray[i][j] = True
tray[0][1] = True

# Print the tray upside down, the entrance is at the bottom
for i in range(10, -1, -1):
    for j in range(11):
        if tray[i][j]:
            print(".", end=" ")
        else:
            print("#", end=" ")
    print()
    
# Init a board with the tray, Alice and Bob, and the exit at (0,1)
board = Board(tray, [alice, bob], Point2D(0, 1))
print(board.robots)

# Move Alice to the exit
print("Moving Alice to the exit")
print(board.moveRobot(0, Point2D(0, 1)))

# Move Bob to the exit
print("Moving Bob to the exit")
print(board.moveRobot(1, Point2D(0, 1)))  

```
